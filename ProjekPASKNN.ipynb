```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for Outcome, d in df.groupby('Outcome'):
    ax.scatter(d['BMI'], d['Age'], label=Outcome)

plt.legend(loc='upper left')
plt.title('Orang Kena Diabetes')
plt.xlabel('Berat Badan Ideal')
plt.ylabel('Umur')
plt.grid(True)
plt.show()
```

```python
import numpy as np

x_train = np.array(df[['BMI', 'Age']])
y_train = np.array(df['Outcome'])

print(f'x_train:\n{x_train}\n')
print(f'y_train:Â {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:Â {y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x_train, y_train)
```

```python
BMI = 33.6
Age = 50
x_new = np.array([BMI, Age]).reshape(1, -1)
x_new
```

```python
y_new = model.predict(x_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

```python
x_test = np.array([[168, 65], [180, 86], [160, 67], [143, 63]])
y_test = lb.transform(np.array(['pria', 'pria', 'wanita', 'wanita']))

print(f'x_test:\n{x_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)


print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)

print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'Recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1 Score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n{cls_report}')
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)

print(f'MCC: {mcc}')
```
